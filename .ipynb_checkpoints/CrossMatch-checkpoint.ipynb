# Crossmatch, por coordenadas

Definimos las SkyCoord de cada catalogo y hacemos un match entre ellos.  
idx = indice del objeto en el catalogo 2  
d2d = la distancia 2d entre ellos  
d3d = distancia 3d entre ellos.  

*¡Hace match con los cercanos! A lo mejor no es el mismo* 

```python
BRA = catalogo_Baldi['RA']
BDEC = catalogo_Baldi['DEC']
Bcoord= SkyCoord(ra= BRA, dec=BDEC, frame="icrs",unit=(u.hourangle, u.deg))


SRA = catalogo_S['RA']
SDEC = catalogo_S['DEC']
Scoord = SkyCoord(ra=SRA, dec=SDEC, frame="icrs", unit=(u.hourangle, u.deg))

LRA = laura['RA']
LDEC = laura['DEC']
Lcoord= SkyCoord(ra= LRA, dec=LDEC, frame="icrs",unit=(u.hourangle, u.deg))

NRA = catalogo_Nagar['_RA']
NDEC = catalogo_Nagar['_DE']
Ncoord= SkyCoord(ra= NRA, dec=NDEC, frame="icrs",unit=(u.deg,u.deg))
```

```python
idx, d2d, d3d = Bcoord.match_to_catalog_sky(Scoord)
```

Entonces tenemos que añadir una condición el que exija que d2d sea 0.  
Buscamos que indice del catalogo 2 (S) lo cumple.

```python
max_sep = 0*u.deg
sep_constraint = d2d == max_sep
print(idx[sep_constraint])
```

Otra opción de hacer lo mismo

```python
max_sep = 0*u.deg
idxa, d2da, d3da = Bcoord.match_to_catalog_3d(Scoord)
sep_constraint = d2da == max_sep
c_matches = Bcoord[sep_constraint]
catalog_matches_coord = catalogo_S[idxa[sep_constraint]]
```

```python
print('\n' 'Índices catálogo S')
print(idxa[sep_constraint])

print('\n' 'Coordenadas B')
print(c_matches.to_string('hmsdms'))

print('\n' 'Tabla con S')

catalog_matches_coord



#catalog_matches_coord.info()
```

# Función final

Funcion: Crear una tabla con los objetos coincidentes(en coordenadas) entre dos catálogos.  
Parámetros:  
    - SkyCoord catálogo 1 (input)  
    - SkyCoord catálogo 2 (input)  
    - Máxima separación (input)  
    - Tabla de salida (input/output)  
    - Catálogo de referencia (input)
    - 5 parametros: columnas en la tabla, segun su nombre en el catálogo de referencia

```python
def crosstable (cat1,cat2,max_sep,t1,tabref,par1,par2,par3,par4,par5):
    idx, d2d, d3d = cat1.match_to_catalog_3d(cat2)
    sep_constraint = d2d < max_sep
    c1_matches = cat1[sep_constraint]
    c2_matches_coord = cat2[idx[sep_constraint]]
    t1 = tabref[idx[sep_constraint]]
    t1.to_pandas()
    t1 = t1[[par1,par2,par3,par4,par5]]
    return t1
```
