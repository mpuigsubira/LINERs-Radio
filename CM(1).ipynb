# Función crear tabla

&nbsp;
Se define una función para crear una tabla a que contenga los elementos coincidentes en dos catálogos y las columnas que nos interesen. 
&nbsp;

 --> **Entrada:**  
 
 Lista de coincidentes (proviene de la funcion crosstable),  
 
 tabla de referencia + los argumentos que nos interesen de la misma (por ahora 3, "name, "RA", Y "DEC"),  
 
 tabla secundaria + los argumentos que correspondasn (N,R,D).   
 
 Se hace general por si nos interesa combinar información de las tablas.

&nbsp;
 --> Debería devolver una tabla que pueda descargar en csv. De momento solo printea, en un futuro, cuando se tenga claro qué guardar, se descargará. 



```python

def crear_tabla(crossmatched, tableref, name, RA, DEC, tablasec, N,R,D):
    RAtab = []
    DECtab = []
    names = []
    coord = []
    p = 0
    for a in tableref[name]:
        a = a.replace(" ","")
        for i in crossmatched:
            i = i.replace(" ","")
            if i == a:
                RAi=tableref[RA][p]
                RAtab.append(RAi)
                DECi= tableref[DEC][p]
                DECtab.append(DECi)
                ID= tableref[name][p]
                names.append(ID)
                
        p = p+1
        
    coord = []
    for n,r,d in zip(names,RAtab, DECtab):
        coordenadas = [n, r,d]
        coord.append(coordenadas)
            
           
    tabla=(pd.DataFrame(coord, columns=['Object', 'RA', 'DEC']))
    #print(p)
    #print(tabla)

    return tabla

```

## Función general, crossmatch por coordenadas

```python
def crosstable (cat1,cat2,max_sep,t1, tref,par1,par2,par3,par4,par5):
    idx, d2d, d3d = cat1.match_to_catalog_3d(cat2)
    sep_constraint = d2d == max_sep
    c1_matches = cat1[sep_constraint]
    c2_matches_coord = cat2[idx[sep_constraint]]
    t1 = tref[idx[sep_constraint]]
    t1.to_pandas()
    t1 = t1[[par1,par2,par3,par4,par5]]
    return t1
```
